```python
# 5 "caixas" vazias para guardar histórico
mem_1 = "vazio"
mem_2 = "vazio"
mem_3 = "vazio"
mem_4 = "vazio"
mem_5 = "vazio"

# Ela serve para o programa lembrar de quem é a vez de receber o próximo resultado.
lugar_da_memoria = 1

# A chave geral do programa.
programa_rodando = "sim"

print("--- CALCULADORA ---")

while programa_rodando == "sim":
    print(" ")
    print("1 - Somar")
    print("2 - Subtrair")
    print("3 - Multiplicar")
    print("4 - Dividir")
    print("5 - Porcentagem")
    print("6 - Ver historico")
    print("7 - Sair")

    opcao = input("Sua escolha: ")

    if opcao == "7":
        print("Tchau!")
        programa_rodando = "nao"

    elif opcao == "6":
        print(" ")
        print("--- HISTORICO ---")
        print("Espaco 1:", mem_1)
        print("Espaco 2:", mem_2)
        print("Espaco 3:", mem_3)
        print("Espaco 4:", mem_4)
        print("Espaco 5:", mem_5)
        print("---------------------")

    elif opcao in ["1", "2", "3", "4", "5"]:
        try:
            # Lemos como texto, trocamos a vírgula por ponto e só depois transformamos em float
            n1_texto = input("Primeiro numero: ").replace(',', '.')
            n1 = float(n1_texto)

            n2_texto = input("Segundo numero: ").replace(',', '.')
            n2 = float(n2_texto)
        except ValueError:
            print("Erro: esta calculadora nao aceita letras!")
            continue

        resultado = 0
        operacao_valida = True

        if opcao == "1":
            resultado = n1 + n2
        elif opcao == "2":
            resultado = n1 - n2
        elif opcao == "3":
            resultado = n1 * n2
        elif opcao == "4":
            if n2 != 0:
                resultado = n1 / n2
            else:
                print("Erro: divisão por zero não é permitida")
                operacao_valida = False
        elif opcao == "5":
            resultado = (n1 * n2) / 100

        if operacao_valida:
            print("Resultado:", resultado)

            if lugar_da_memoria == 1:
                mem_1 = resultado
                lugar_da_memoria = 2
            elif lugar_da_memoria == 2:
                mem_2 = resultado
                lugar_da_memoria = 3
            elif lugar_da_memoria == 3:
                mem_3 = resultado
                lugar_da_memoria = 4
            elif lugar_da_memoria == 4:
                mem_4 = resultado
                lugar_da_memoria = 5
            elif lugar_da_memoria == 5:
                mem_5 = resultado
                lugar_da_memoria = 1
```
